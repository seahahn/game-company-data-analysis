## 3. ê¸°ì´ˆ ë¯¸ë¶„

ë‹¤ìŒ í•¨ìˆ˜ë¥¼ ë¯¸ë¶„í•œ ë„í•¨ìˆ˜ë¥¼ êµ¬í•œ í›„ì— num_one í•¨ìˆ˜ë¥¼ ì™„ì„±í•˜ì„¸ìš”. 
$f(x) = 4x^5 + 9x^3 + ex^2 - 7$

- num_one í•¨ìˆ˜ë¥¼ xì— ì ìš©í•œ í›„ì— **derivative_one**ì´ë¼ëŠ” ì¹¼ëŸ¼ìœ¼ë¡œ df_irisì— ì €ì¥í•©ë‹ˆë‹¤. (Hint : df.apply)
- derivative_one ì¹¼ëŸ¼ì˜ í•©ì„ `res2`ì— ì €ì¥í•©ë‹ˆë‹¤ (ì†Œìˆ˜ì  ì„¸ìë¦¬ê¹Œì§€)

```python
import numpy as np

e = np.exp(1)

# ë„í•¨ìˆ˜ êµ¬í•˜ê¸°
def num_one(x):
  #20x^4 + 27x^2 + 2ex
  return 20*(x**4) + 27*(x**2) + 2*e*x

# derivative_one ì»¬ëŸ¼ ì €ì¥
df_iris['derivative_one'] = df_iris['x'].apply(num_one)
df_iris
```

```python
# derivative_one ì»¬ëŸ¼ í•©ì„ res2ì— ì €ì¥
res2 = round(sum(df_iris['derivative_one']), 3)
res2
```

- ì´ë²ˆì—ëŠ” scipyì˜ [derivative](https://docs.scipy.org/doc/scipy/reference/generated/scipy.misc.derivative.html)ë¥¼ í™œìš©í•˜ì—¬ **derivative_two** ì¹¼ëŸ¼ìœ¼ë¡œ ì €ì¥í•˜ì„¸ìš”.
  - dxëŠ” 1e-7ë¡œ ì„¤ì •í•©ë‹ˆë‹¤.
- derivative_two ì¹¼ëŸ¼ì˜ í•©ì„ `res3`ì— ì €ì¥í•©ë‹ˆë‹¤ (ì†Œìˆ˜ì  ì„¸ìë¦¬ê¹Œì§€)

```python
from scipy.misc import derivative

# ë¯¸ë¶„í•˜ê¸° ì „ì˜ í•¨ìˆ˜
def f(x):
    return 4*(x**5) + 9*(x**3) + e*(x**2) - 7

# scipyì˜ derivativeë¥¼ í™œìš©í•˜ì—¬ derivative_two ì¹¼ëŸ¼ìœ¼ë¡œ ì €ì¥
df_iris['derivative_two'] = derivative(f, df_iris['x'], dx=1e-7)
df_iris
```

```python
# derivative_two ì»¬ëŸ¼ì˜ í•©ì„ res3ì— ì €ì¥
res3 = round(sum(df_iris['derivative_two']), 3)
res3
```
