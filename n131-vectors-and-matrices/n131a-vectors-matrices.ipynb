## 2. Norms

ì£¼ì–´ì§„ ê°’ $x_0, x_1, ... x_n$ ë¥¼ elementë¡œ ê°–ëŠ” ë²¡í„° $v$ì— ëŒ€í•˜ì—¬ 

$L_1$ Norm ì˜ ì •ì˜ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤. 

$|v| = |x_0| + |x_1| + ... + |x_n|$

$L_2$ Norm ì˜ ì •ì˜ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.

$||v|| = \sqrt {{x_0}^2 + {x_1}^2 + ... + {x_n}^2}$

ì´ë¥¼ ì°¸ì¡°í•˜ì—¬ ì£¼ì–´ì§„ ë²¡í„°ì— ëŒ€í•´ì„œ $L_2$ Norm ì˜ ê°’ê³¼ $L_1$ Norm ì˜ ê°’ì˜ ì°¨ì´(**ì ˆëŒ€ê°’**)ë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“œì„¸ìš”.



```python
def NormDif(v): # v = np.array()
  l_1 = sum(np.abs(v))
  l_2 = np.sqrt(sum(np.square(v)))
  return np.abs(l_2 - l_1)
```

```python
ex = np.array([100, 500, 250, 50])
NormDif(ex)
```

## 3. Errors



- 2ê°œì˜ ë²¡í„°ì™€, errorì˜ íƒ€ì…ì„ ì…ë ¥ë°›ì•„ ê·¸ ê²°ê³¼ê°’ì„ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.

- ì—¬ê¸°ì„œ errorì˜ íƒ€ì…ì€ `MSE`, `MAE` 2ê°€ì§€ ì…ë‹ˆë‹¤.



```python
def error(x, y, type) :
  if type == 'MSE':
    result = sum(np.square(x-y))/len(x)
  elif type == 'MAE':
    result = sum(np.abs(x-y))/len(x)
  else:
    result = 'type error'
  return result
```

```python
ex1 = np.array([0, 50, 50, 0])
ex2 = np.array([0, 0, 100, 0])
error(ex1, ex2, 'MSE'), error(ex1, ex2, 'MAE')
```

```python
def myInverse(m) : # m = np.array()
  try:
    det = np.linalg.det(m)
    return np.linalg.inv(m)
  except:
    return -1
```

# ğŸ”¥ ë„ì „ê³¼ì œ 



ë‹¤ìŒ ë§í¬ì˜ ë‚´ìš©ì„ ì°¸ì¡°í•˜ì—¬ **Cramer's rule**ì„ ì‚¬ìš©í•´ $x_1$, $x_2$, $x_3$ì˜ ê°’ì„ êµ¬í•˜ì„¸ìš”.

$x_1$ + $2x_3$ = 6

$-3x_1$ + $4x_2$ + $6x_3$ = 30

$-x_1$ $-2x_2$ + $3x_3$ = 8



```python
# ì—°ë¦½ë°©ì •ì‹ì„ í–‰ë ¬ë¡œ ë§Œë“¤ê¸°
A = np.array([
     [1, 0, 2],
     [-3, 4, 6],
     [-1, -2, 3]
])

B = np.array([
     [6],
     [30],
     [8]
])
```

```python
# í¬ë¼ë©”ë¥´ ê³µì‹ í•¨ìˆ˜ë¡œ êµ¬í˜„
def cramer(A, B):
  x_arr = [] # ë¯¸ì§€ìˆ˜ ê²°ê³¼ê°’ ë„£ì„ ë¦¬ìŠ¤íŠ¸
  det_a = np.linalg.det(A) # ì¢Œë³€ì˜ ë¯¸ì§€ìˆ˜ì— ë¶™ì–´ ìˆëŠ” í–‰ë ¬ Aì˜ í–‰ë ¬ì‹ êµ¬í•˜ê¸°
  
  for i in range(len(B)): # ë¯¸ì§€ìˆ˜ ê°œìˆ˜ = í–‰ë ¬ Bì˜ ê¸¸ì´ ì´ë¯€ë¡œ Bì˜ ê¸¸ì´ë§Œí¼ ë°˜ë³µ
    n = A.copy() # í–‰ë ¬ A ê·¸ëŒ€ë¡œ ê°€ì ¸ì˜¤ê¸°(ê° ë¯¸ì§€ìˆ˜ì— í•´ë‹¹í•˜ëŠ” ì—´ ê°’ ë°”ê¾¸ê¸° ìœ„í•¨)
    n[:, i] = B[:, 0] # ë¯¸ì§€ìˆ˜ ìœ„ì¹˜ì— í•´ë‹¹ë˜ëŠ” ì—´ ê°’ ë°”ê¾¸ê¸°
    det_sub = np.linalg.det(n) # ì—´ ê°’ ë°”ê¾¼ í–‰ë ¬ì˜ í–‰ë ¬ì‹ êµ¬í•˜ê¸°
    result = det_sub / det_a
    x_arr.append(result)
  
  return x_arr
```

```python
cramer(A, B)
```
